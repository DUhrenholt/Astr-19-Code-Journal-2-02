```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,1,100)

def b(x):
    return np.sin(x)
def d(x):
    return np.cos(x)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, np.sin(x))
ax1.set_ylabel('sin(x)')
ax1.set_xlabel('x')
ax2.plot(x, np.cos(x))
ax2.set_ylabel('cos(x)')
ax2.set_xlabel('x')
plt.subplots_adjust(wspace=0.5)
plt.savefig('sin(x)_vs_x_AND_cos(x)_vs_x.pdf')
```
